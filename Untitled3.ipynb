```python
!pip install english_words
from english_words import get_english_words_set
words = get_english_words_set(['web2'], lower=True)

wordlist = list(words)
wordlist_length = len(wordlist)

print("First ten words in wordlist:", wordlist[:10])

alpha_list = sorted(wordlist)
print("First ten words in alpha_list:", alpha_list[:10])

words_starting_with_a = sum(1 for word in wordlist if word.startswith('a'))

words_with_q = sum(1 for word in wordlist if 'q' in word)
fraction_without_e = sum(1 for word in wordlist if 'e' not in word) / wordlist_length

import numpy as np
wordlist_array = np.array(wordlist)
alpha_list_array = np.array(alpha_list)

alphabet = [chr(i) for i in range(ord('a'), ord('z')+1)]

import matplotlib.pyplot as plt
from collections import Counter

letter_counts = Counter(word[0] for word in wordlist)
letters = [letter for letter in alphabet if letter in letter_counts]
counts = [letter_counts[letter] for letter in letters]

plt.bar(letters, counts)
plt.xlabel('Letters')
plt.ylabel('Number of words')
plt.title('Number of words starting with each letter')
plt.show()

six_longest_words = sorted(wordlist, key=len, reverse=True)[:6]
print("Six longest words:", six_longest_words)

word_lengths = [len(word) for word in wordlist]
plt.hist(word_lengths, bins=range(min(word_lengths), max(word_lengths) + 1), align='left')
plt.xlabel('Word Length')
plt.ylabel('Frequency')
plt.title('Histogram of word lengths')
plt.show()

most_common_length = max(set(word_lengths), key=word_lengths.count)
print("Most common length of a word:", most_common_length)
```
