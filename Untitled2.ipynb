```python
import matplotlib.pyplot as plt
import requests
import numpy as np
from PIL import Image

def red_filter(photo):
    red_only = np.copy(photo)
    red_only[:, :, 1:] = 0
    return red_only
def green_filter(photo):
    green_only = np.copy(photo)
    green_only[:, :, [0, 2]] = 0
    return green_only
def blue_filter(photo):
    blue_only = np.copy(photo)
    blue_only[:, :, :2] = 0
    return blue_only

url = input("Enter the URL of a .jpg file: ")

response = requests.get(url)
photo = np.asarray(Image.open(requests.get(url, stream=True).raw))

photo = np.array(photo)

red_filtered = red_filter(photo)
green_filtered = green_filter(photo)
blue_filtered = blue_filter(photo)

rg_filtered = np.copy(photo)
rg_filtered[:, :, 2] = 0
rb_filtered = np.copy(photo)
rb_filtered[:, :, 1] = 0
gb_filtered = np.copy(photo)
gb_filtered[:, :, 0] = 0

plt.figure(figsize=(12, 8))
plt.subplot(2, 3, 1)
plt.imshow(photo)
plt.title('Original')
plt.subplot(2, 3, 2)
plt.imshow(red_filtered)
plt.title('Red Filtered')
plt.subplot(2, 3, 3)
plt.imshow(green_filtered)
plt.title('Green Filtered')
plt.subplot(2, 3, 4)
plt.imshow(blue_filtered)
plt.title('Blue Filtered')
plt.subplot(2, 3, 5)
plt.imshow(rg_filtered)
plt.title('RG Filtered')
plt.subplot(2, 3, 6)
plt.imshow(rb_filtered)
plt.title('RB Filtered')
plt.tight_layout()
plt.show()
is_identical = np.array_equal(photo, red_filtered + green_filtered + blue_filtered)

if is_identical:
    print("Original image is identical to the combined filtered images.")
else:
    print("Original image is not identical to the combined filtered images.")

print("holy crap this took me a minute")
```
